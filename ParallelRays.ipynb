## Checking the Math
---
### Let's check the actual angular span, as we can't trust our eyes.

```python
def ray_angles(radius, x, y=0):
    # Slopes of tangent lines
    m_top = (y + radius) / x #Top of Disc to Solar Center
    m_bottom = (y - radius) / x #Bottom of Disc to Solar Center
    m_center = y/x #Straight Line to Center (slope = 0)

    # Angles of tangent lines
    theta_top = np.arctan(m_top) #Angle to Top of Disc
    theta_bottom = np.arctan(m_bottom) #Angle to Bottom of Disc
    theta_center = np.arctan(m_center) #Center to Center

    # Angular span between tangent rays
    delta_theta = theta_top - theta_bottom #Full angle from Solar Center to Edges of Disc
    delta_theta_top_center = theta_top - theta_center #Top of Disc Angle
    delta_theta_center_bottom = theta_center - theta_bottom #Bottom of Disc Angle
 
    # Convert to degrees
    theta_top_deg = np.degrees(theta_top) #Angle to Top of Disc
    theta_bottom_deg = np.degrees(theta_bottom) #Angle to Bottom of Disc
    theta_center_deg = np.degrees(theta_center)
    
    delta_theta_deg = np.degrees(delta_theta)
    delta_top_center_deg = np.degrees(delta_theta_top_center)
    delta_center_bottom_deg = np.degrees(delta_theta_center_bottom)

    # Output results
    print(f"Solar Center to Upper Disc Edge: {theta_top_deg:.8f} degrees"),
    print(f"Solar Center to Bottom Disc Edge: {theta_bottom_deg:.8f} degrees"),
    print(f'Solar Center to Disc Center: {theta_center_deg:.8f} degrees'),
    print(f'Top to Center Span: {delta_top_center_deg:.8f} degrees'),
    print(f'Center to Bottom Span: {delta_center_bottom_deg:.8f} degrees'),
    print(f"Total Angular Span Across Disc: {delta_theta_deg:.8f} degrees")

    return
```

```python
ray_angles(earth_radius, distance_to_earth)
```

## How does this translate to a tilted Earth?

```python
# Constants
TSI = 1367  # Total Solar Irradiance in W/m2
tilt = -23.45  # Earth's axial tilt in degrees
declination = 23.45 # solar declination
tilt_rad = np.deg2rad(tilt)
lon = np.linspace(-180, 180, 360)  # Longitude
lat = np.linspace(-90, 90, 180)  # Latitude
lon_grid, lat_grid = np.meshgrid(lon, lat)

# Subsolar latitudes for the events
events = {
    "Equinox": 0,
    "June Solstice": declination,
    "December Solstice": -declination,
}
```

```python
# Function to *estimate* DNI (direct beam) for Local Solar Noon, based on spherical geometry
# This can be done in more detail based on specific hours of day, maybe we'll look at Local Solar Midnight too?
def estimate_dni(lat_grid, lon_grid, subsolar_lat):
    # Solar zenith angle in degrees
    solar_zenith_angle = np.abs(lat_grid - subsolar_lat)
    # Cosine law for irradiance
    dni = TSI * np.maximum(0, np.cos(np.radians(solar_zenith_angle)))
    return dni
```

```python
cmap = cm.cividis  # Colormap for shading

Esubsolar_lat = 0  #Equator
Edni = estimate_dni(lat_grid, lon_grid, Esubsolar_lat)
# Normalize DNI for shading
Enorm = Normalize(vmin=0, vmax=Edni.max())  # Normalize between 0 and TSI
Ecolors = cmap(Enorm(Edni))

# June Solstice with Tilt
Jsubsolar_lat = declination  # 23.5 (North)
Jdni = estimate_dni(lat_grid, lon_grid, Jsubsolar_lat)
# Normalize DNI for shading
Jnorm = Normalize(vmin=0, vmax=Jdni.max())  # Normalize between 0 and TSI
Jcolors = cmap(Jnorm(Jdni))

# December Solstice with Tilt
Dsubsolar_lat = -declination  # -23.45 (South)
Ddni = estimate_dni(lat_grid, lon_grid, Dsubsolar_lat)
# Normalize DNI for shading
Dnorm = Normalize(vmin=0, vmax=Ddni.max())  # Normalize between 0 and TSI
Dcolors = cmap(Dnorm(Ddni))
```

```python
# Plot Seasons with Plate Carree Projection
fig, ax = plt.subplots(figsize=(12, 8), subplot_kw={'projection': ccrs.PlateCarree()})
ax.set_global()
ax.coastlines()
ax.set_title("Equinox")
gl = ax.gridlines(draw_labels={"bottom": "x", "left":"y"}, xlocs=[-180, -90, 0, 90, 180], ylocs=[-90, -45, 0, 45, 90])
    
# Plot
im = ax.pcolormesh(lon, lat, Edni, transform=ccrs.PlateCarree(), cmap='cividis', shading='auto')
    
# Highlight subsolar latitude
ax.plot([-180, 180], [0, 0], color='black', linestyle='-', transform=ccrs.PlateCarree(), label="Equator")
ax.plot([-180, 180], [0, 0], color='red', linestyle='--', transform=ccrs.PlateCarree(), label="Subsolar Latitude")
    
ax.legend(loc='lower left')

# Colorbar
cbar = fig.colorbar(im, ax=ax, orientation='horizontal', use_gridspec = True, pad = 0.075)
cbar.set_label("Estimated DNI at Local Solar Noon (W/m)")
plt.savefig('EquiFlat.png', bbox_inches='tight')
plt.show()
```

```python
# Plot Equinox with Tilt
# Create the figure and axis
fig, ax = plt.subplots(figsize=(10, 10))

# Earth center
earth_center = (distance_to_earth, 0)

# Draw Earth's outline
earth_outline = plt.Circle(earth_center, earth_radius, facecolor='none', edgecolor='black', lw=1.5, zorder=4)
ax.add_patch(earth_outline)

# Draw Earth's axial tilt
ax.plot(
    [distance_to_earth - earth_radius, distance_to_earth + earth_radius],
    [earth_center[1] - (earth_radius / np.tan(-tilt_rad)), earth_center[1] + (earth_radius / np.tan(-tilt_rad))],
    color='black', linestyle='--', label=f"Axial Tilt ({-tilt})", zorder=5
)


# Scale the latitude/longitude grid to match the Earth's size
# Longitude and latitude grid need to map properly to Earth's size
lat_rad = np.radians(lat_grid)
lon_rad = np.radians(lon_grid)

# Calculate the x and y coordinates for the mesh on the Earth's surface
x = distance_to_earth + earth_radius * np.cos(lat_rad) * np.cos(lon_rad)
y = earth_radius * np.sin(lat_rad)

# Temporary x for tranlastion
x_temp = x - distance_to_earth

# Rotate mesh to match axial tilt
x_rot = x_temp * np.cos(-tilt_rad) + y * np.sin(-tilt_rad) + distance_to_earth
y_rot = -x_temp * np.sin(-tilt_rad) + y * np.cos(-tilt_rad)


# Shade the Earth with DNI values using pcolormesh
# Use longitude and latitude grid for the mesh
mesh = ax.pcolormesh(x_rot, y_rot, Edni, cmap=cmap, norm=Enorm, shading='auto', zorder=3)

# Labels, legend, and adjustments
#ax.set_xlim((distance_to_earth - 2 * earth_radius, distance_to_earth + 2 * earth_radius))
#ax.set_ylim(-2 * earth_radius, 2 * earth_radius)



# Add colorbar
sm = cm.ScalarMappable(cmap=cmap, norm=Enorm)
sm.set_array([])
cbar = plt.colorbar(sm, ax=ax, orientation='vertical', fraction=0.025, pad=0.1)
cbar.set_label("Estimated DNI at Local Solar Noon (W/m)")

# Set axis limits to zoom in around the Earth
ax.set_xlim((distance_to_earth - (earth_radius + 0.05)), (distance_to_earth + (earth_radius + 0.05)))  # Focus on Earth region
ax.set_ylim(-0.08, 0.08)  # Zoom in on the area around Earth
ax.set_aspect('equal', 'box')
ax.minorticks_on()

# Labels, legend, and adjustments
ax.set_xlabel("Sun is in front of screen!")
ax.set_ylabel("Distance from Solar Center (100,000 km)")
ax.grid(which='both', linestyle='--', alpha=0.7) # Add grid for comparison to parallel
ax.legend(loc="upper left")

#Hide x-labels
ax.set_xticklabels([])  # Sun is in front of the screen!!

# Add notes
ax.text(distance_to_earth + 0.09, 0, 'Earth\n(Actual Size)', ha='center', va='center', fontsize=12, color='black')

plt.title("Zoomed-In View: Parallel Rays at TOA with DNI Shading\nEquinox")
plt.savefig('Equinox.png', bbox_inches='tight')
plt.show()
```
